### __1.__ Hacer una función que crea un dataframe a partir de los atributos de los datos de entrada examen3.xlsx, los cuales son:

#### __a.__ Nombre de la columna del dataframe de entrada. 
#### __b.__ Tipo de dato (incluido: int, float, string, datetime, boolean). 
#### __c.__ Tipo de variable (Continua, numérica discreta, categórica). 
#### __d.__ Cantidad de valores únicos. 
#### __e.__ Cantidad de valores nulos. 
#### __f.__ Valor que más se repite. 
#### __g.__ Frecuencia del valor que más se repite. 
#### __h.__ Valor máximo. 
#### __i.__ Valor mínimo.

```
# Se importan las librerias de google colab
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

#En python con el comando "def"" realizas funciones
def crear_dataframe(ruta = "examen3.xlsx"):
    
  # Se lee el archivo EXCEL con la libreria pandas 
  # Se lee como un dataframe por defecto
  data = pd.read_excel(ruta)
  print("Tipo de archivo excel: " + str(type(data)))
  print()

  #Nombres de las columnas
  names_columns = data.columns[:]

  for name in names_columns:

    print(str(name) + ": " + str(type(data[name].values[0])))

    ### Valores numericos int64 y float64 ######################################
    if str(type(data[name].values[0])) == "<class 'numpy.int64'>" or str(type(data[name].values[0])) == "<class 'numpy.float64'>":
      
      #Cuenta la cantidad de valores unicos
      count_of_uniques = data[name].nunique()

      #Cuenta la cantidad de valores nulos
      count_of_nan_values = data[name].isna().sum()

      #Valores mas frecuentes
      frecuent_values = data[name].value_counts().head(1)

      #Valor maximo
      maxium_value = data[name].max()

      #Valor minimo
      min_value = data[name].min()
      
      print(name + ", Tipo: " + str(type(data[name].values[0])) + 
            ", Variable continua" + ", Valores únicos: " + str(count_of_uniques) +
            ", Valores nulos: " + str(count_of_nan_values) + 
            ", Valor mas frecuente y sus repeticiones: " + str(frecuent_values) +
            ", el valor máximo es: " + str(maxium_value)+
            ", el valor mínimo es: " + str(min_value))
      
      print()

    ### Valores string #########################################################
    elif str(type(data[name].values[0])) == "<class 'str'>":

      #Cuenta la cantidad de valores unicos
      count_of_uniques = data[name].nunique()

      #Cuenta la cantidad de valores nulos
      count_of_nan_values = data[name].isna().sum()
      
      #Valores mas frecuentes
      frecuent_values = data[name].value_counts().head(1)
      
      print(name + ", Tipo: " + str(type(data[name].values[0])) + 
            ", Variable categórica" + ", Valores únicos: " + str(count_of_uniques) +
            ", Valores nulos: " + str(count_of_nan_values) + 
            ", Valor mas frecuente y sus repeticiones: " + str(frecuent_values))
      print()

  return data
```

```
from datetime import datetime, timedelta, date, time

def corregir_datos(dataframe):

  dataframe["programa"].replace({"Ingeniería Electórnica": "Ingeniería Electrónica",
                                 "Ingeniería Industial": "Ingeniería Industrial"}, inplace=True)
  
  dataframe["fallas"].replace({"nada": 0, "ninguna": 0, "no": 0}, inplace=True)
  dataframe["dedicacion"].replace({"tc": "TC"}, inplace=True)
  dataframe["horaExamen"] = pd.to_datetime(dataframe['horaExamen']).dt.time
  dataframe["inicio"] = pd.to_datetime(dataframe['inicio']).dt.time
  dataframe["final"] = pd.to_datetime(dataframe['final']).dt.time

  dataframe.info()

  print(type(dataframe.horaExamen[0]))
  print(type(dataframe.inicio[0]))
  print(type(dataframe.final[0]))
  
  return dataframe
```

En la columna de "Programas" se encontraron programas con nombres mal escritos, por ejemplo tanto como existe el programa de Ingeniería Electrónica tambien existe Ingeniería Electórnica con la letra R ubicada en diferente orden, como tambien existe el caso en el que se presenta Ingeniería Industrial e Ingeniería Industial sin la letra R.

En las columnas de inicio, final y hora examen se tenían lecturas de datos tipo string, cuando el valor real debería pertenecer a un formato datetime, por lo que se realizó una conversión entre los datos tipo string a tipo datetime.

Tomando en cuenta que los valores de la columna nota tiene valores numéricos y valores nulos coincidentes con los valores nulos asociados a las filas exactas de otras columnas, se procedió a realizar la eliminación de filas basado en los valores nulos de la columna nota.

#### __Tipos, valores únicos y nulos__

Se creó una función que permite realizar el análisis de los datos para detectar tipos de variables, valores únicos y valores nulos en cada una de las columnas de la base de datos.

```
# funcion revisar tipo de datos y análisis

def datos_y_analisis(data):
  print('Valores nulos por columna: shape', data.shape)
  print('Columnas \t tipo\túnicos\tnulos')
  print('----------------------------------------')
  for col in data.columns.to_list():
    print('%s:    \t%s\t%3d/%d\t%2d\t'%(col, data[col].dtype, 
                                        len(data[col].unique()),
                                        len(data),
                                        len(data[data[col].isnull()])))
```

```
#Ejecutamos la funcion sobre los datos reparados
datos_y_analisis(data_reparada)
```

Se modifican los valores de las filas con respecto a la columna notas usando el promedio de cada una de sus columnas con valores nulos, como lo son: inicio, final, nota, p1, p2, p3, p4, p5, p6, p7, p8, p9 y p10.

```
#Busca en que posiciones o indices se encuentran los valores nulos en la columna nota
data_reparada_missing_ind = data_reparada[data_reparada.nota.isnull()].index

#Muestra la data en las filas donde se encontraron valores nulos
data_reparada.loc[data_reparada_missing_ind]
```

Se crea la fecha de inicio y final del examen en tiempo cero de cada uno de los estudiantes que no presentaron el examen.

```
cero_horas = time()

data_reparada.loc[data_reparada_missing_ind, 'inicio'] = cero_horas
data_reparada.loc[data_reparada_missing_ind, 'final'] = cero_horas
```

Se crea una nueva columna llamada tiempo de desarrollo, esta columna indicara el tiempo demorado en realizar cada uno de los examenes tomando en cuenta la diferencia horaria de inicio del examen y final del examen.

```
#Se agrega la columna tiempo de desarrollo del examen
data_reparada['Tiempo de desarrollo'] = data_reparada['IDE']

for i in range(len(data_reparada['final'])):
  tiempo = datetime.combine(date.today(), data_reparada['final'][i]) - datetime.combine(date.today(), data_reparada['inicio'][i])
  data_reparada['Tiempo de desarrollo'][i] = int(tiempo.seconds/60)
```

Se reemplazaron las notas de cada uno de los puntos del examen con la nota de cero dado que los estudiantes no se presentaron al examen.

```
data_reparada.loc[data_reparada_missing_ind, 'p1'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p2'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p3'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p4'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p5'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p6'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p7'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p8'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p9'] = 0
data_reparada.loc[data_reparada_missing_ind, 'p10'] = 0

#La nota final es la sumatoria del promedio obtenido en cada uno de los puntos del examen
data_reparada.loc[data_reparada_missing_ind, 'nota'] = 0

#Muestra la data modificada
data_reparada.loc[data_reparada_missing_ind]
```

Nota_total = p1 + p2 + p3 + p4 + p5 + __p6__ + p7 + p8 + p9 + 10

__Despejando p6__


Nota_total - (p1 + p2 + p3 + p4 + p5 + p7 + p8 + p9 + 10) = __p6__ 

__p6__ = Nota_total - (p1 + p2 + p3 + p4 + p5 + p7 + p8 + p9 + 10)

Debido a que la división se realiza de fallaP sobre fallas se encuentran valores de cero por lo que una división por cero arrojaria valores nulos, por ende se procedió a realizar un ciclo for que hiciera la división entre los valores tomando los valores nulos como ceros.

```
#Valor de fallasP sobre fallas
Valor_fallasP = []

for i in range(len(data_reparada['fallaP'])):

  #División
  resultado = data_reparada['fallaP'][i]/data_reparada['fallas'][i]
  
  if(str(resultado) != 'nan'):
    Valor_fallasP.append(resultado)

  else:
    Valor_fallasP.append(0)
```

```
profesores = dataframe_new['profesor'].value_counts(ascending=False,normalize=True).index
```

```
#Listas de datos
lista_profesores = []
lista_promedios_profesores = []
```

```
for profesor in profesores:

  #Busca en que posiciones o indices se encuentran los profesores
  dataframe_new_missing_ind = dataframe_new[dataframe_new.profesor == profesor].index

  #Muestra la data en las filas donde se encontraron los profesores y halla el promedio
  promedio_de_los_estudiantes = dataframe_new.loc[dataframe_new_missing_ind].prom.mean()

  lista_profesores.append(profesor)
  lista_promedios_profesores.append(promedio_de_los_estudiantes)

  print("El promedio de los estudiantes a los que le dicto el/la profesor/a " + profesor + " fue de: " + str(round(promedio_de_los_estudiantes,2)))
```
